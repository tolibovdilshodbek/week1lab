# Problem 2 week 1

```
study_hours = [1, 2, 3, 4, 5]
exam_scores = [50, 55, 65, 70, 80]


print("\nStudy Hours and Exam Scores:")
for hours, score in zip(study_hours, exam_scores):
    print(hours, score)


print("\nInput → Output:")
for hours, score in zip(study_hours, exam_scores):
    print(f"Input: {hours} hours → Output: {score} marks")
```

# Problem 3 week 1

```
scores = [45, 67, 82, 59]

def pass_fail(score):
    if score >= 60:
        return "Pass"
    else:
        return "Fail"

print("Highest score:", max(scores))
print("Percentage:", sum(scores) / len(scores))

for score in scores:
    print("Score:", score, "->", pass_fail(score))
```

# Problem 4 week 1

```
sizes = [500, 800, 1000, 1200]
prices = [50, 80, 100, 130]

for i in range(len(sizes)):
    print("Size:", sizes[i], "-> Price:", prices[i])

positive_trend = True

for i in range(len(prices) - 1):
    if prices[i] >= prices[i + 1]:
        positive_trend = False
        break

if positive_trend:
    print("Positive trend")
```

# Problem 6 week 1

```
scores = [35, 58, 62, 74, 89, 45]

for score in scores:
    if score < 50:
        category = "Low"
    elif score <= 70:
        category = "Medium"
    else:
        category = "High"

    print(f"Score: {score}, Category: {category}")
```

# Problem 7 week 1

```
ages = [18, 20, 22, 24, 26]
heights = [160, 165, 170, 175, 180]

for age, height in zip(ages, heights):
    print(f"Age: {age}, Height: {height}")

average_age = sum(ages) / len(ages)
print("Average age:", average_age)

average_height = sum(heights) / len(heights)
print("Average height:", average_height)

print("Height appears to increase as age increases in this dataset.")
```
